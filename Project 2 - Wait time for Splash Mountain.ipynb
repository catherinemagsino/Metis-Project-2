```python
import pandas as pd
```

```python
Ride_Splash_Mtn=pd.read_csv('splash_mountain.csv')
```

```python
Ride_Splash_Mtn
```

```python
Ride_Splash_Mtn['date'] = pd.to_datetime(Ride_Splash_Mtn['date'])
```

```python
Ride_Splash_Mtn['datetime'] = pd.to_datetime(Ride_Splash_Mtn['datetime'])
```

```python
Ride_Splash_Mtn['Date'] = Ride_Splash_Mtn['date'].dt.strftime('%Y-%m-%d')
```

```python
#Ride_Splash_Mtn['Date'] = pd.to_datetime(Ride_Splash_Mtn['Date'])
```

```python
Ride_Splash_Mtn['SPOSTMIN'].fillna(0,inplace=True)
```

```python
Ride_Splash_Mtn['SACTMIN'].fillna(0,inplace=True)
```

```python
Ride_Splash_Mtn['Wait_Time']=Ride_Splash_Mtn['SPOSTMIN']+Ride_Splash_Mtn['SACTMIN']
```

```python
Ride_Splash_Mtn['Hour'] = Ride_Splash_Mtn['datetime'].dt.hour
```

```python
Ride_Splash_Mtn['Day_of_the_Week'] = Ride_Splash_Mtn['datetime'].dt.weekday_name
```

```python
Ride_Splash_Mtn
```

```python
Ride_Splash_Mtn['Month'] = Ride_Splash_Mtn['datetime'].dt.month
```

```python
Ride_Splash_Mtn
```

```python
Ride_Splash_Mtn['Date_and_Hour']=Ride_Splash_Mtn['Date'].map(str)+'_'+Ride_Splash_Mtn['Hour'].map(str)
```

```python
Ride_Splash_Mtn
```

```python
Ride_Splash_Mtn.to_csv("Ride_Splash_Mtn.csv")
```

```python
Ride_Splash_Mtn_May_Aug_2018=pd.read_csv('Ride_Splash_Mtn_May_Aug_2018.csv')
```

```python
Master_df=pd.merge(Ride_Splash_Mtn_May_Aug_2018,WeatherU_May_Aug_2018,on="Date_and_Hour",how="left")
```

```python
Master_df
```

```python
Master_df.info()
```

```python
Master_df['Date']=pd.to_datetime(Master_df['Date_x'])
```

```python
Master_df['Hour']=Master_df['Hour_x']
```

```python
Variables_for_Master=['Date','datetime','Wait_Time','Month','Day_of_the_Week','Hour','Temp_(F)','Dew_Point_(F)','Humidity_(%)','Wind_Speed_(mph)','Wind_Gust_(mph)','Pressure_(in)','Precip_(in)']
```

```python
Disney_Wait_df=Master_df.loc[:,Variables_for_Master]
```

```python
Disney_Wait_df
```

```python
Disney_Wait_df.info()
```

```python
Disney_Wait_df_with_dummies=pd.get_dummies(Disney_Wait_df,columns=['Day_of_the_Week','Hour'])
```

```python
Disney_Wait_df_with_dummies
```

```python
Disney_Wait_no_outliers=Disney_Wait_df_with_dummies.loc[Disney_Wait_df_with_dummies['Wait_Time']!=-999,:]
```

```python
Disney_Wait_no_outliers.to_csv('df_Splash_Mtn.csv',index=False)
```

```python
Disney_Wait_no_outliers
```

```python
Disney_Wait_no_outliers2=Disney_Wait_no_outliers.loc[Disney_Wait_no_outliers['Temp_(F)']!=0,:]
```
